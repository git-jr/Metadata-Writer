# TensorFlow Lite Metadata Writer API

## Prerequisites

Install the TensorFlow Lite Support package and add `metadata_writers` from source (required for Python 3.12+, since the pip wheel omits this subpackage).

```
import subprocess, sys, os, shutil, importlib.util

# ── Step 1: Install flatbuffers (only external dep needed) ───────────────────
print('[1] Installing flatbuffers...')
r = subprocess.run(['pip', 'install', 'flatbuffers', '-q'], capture_output=True, text=True)
if r.returncode != 0:
    raise SystemExit(f'pip failed: {r.stderr}')
print('    OK')

# ── Step 2: Sparse-clone tflite-support source ───────────────────────────────
SRC = '/tmp/ts-src'
if not os.path.exists(SRC):
    print('[2] Cloning source...')
    subprocess.run(
        ['git', 'clone', '--depth=1', '--filter=blob:none', '--sparse',
         'https://github.com/tensorflow/tflite-support', SRC],
        capture_output=True, check=True
    )
    subprocess.run(
        ['git', '-C', SRC, 'sparse-checkout', 'set', 'tensorflow_lite_support/metadata'],
        capture_output=True, check=True
    )
    print('    OK')
else:
    print('[2] Source already present')

# ── Step 3: Build pure-Python tflite_support shim ────────────────────────────
SHIM     = '/tmp/tflite_shim'
PKG      = os.path.join(SHIM, 'tflite_support')
META_SRC = f'{SRC}/tensorflow_lite_support/metadata'
os.makedirs(PKG, exist_ok=True)

# __init__.py — wire tflite_support.flatbuffers → real flatbuffers
open(os.path.join(PKG, '__init__.py'), 'w').write(
    'import sys, flatbuffers as _fb\n'
    'sys.modules.setdefault("tflite_support.flatbuffers", _fb)\n'
)

# Schema files (pure-Python FlatBuffers, present in source)
for fname in ['metadata_schema_py_generated.py', 'schema_py_generated.py']:
    src = os.path.join(META_SRC, fname)
    if os.path.exists(src):
        shutil.copy(src, os.path.join(PKG, fname))
        print(f'    schema: {fname}')
    else:
        print(f'    WARNING: {fname} not in source')

# metadata.py (MetadataDisplayer — pure-Python FlatBuffers reader)
meta_py = os.path.join(META_SRC, 'python', 'metadata.py')
if os.path.exists(meta_py):
    txt = open(meta_py).read()
    txt = txt.replace('tensorflow_lite_support', 'tflite_support')
    open(os.path.join(PKG, 'metadata.py'), 'w').write(txt)
    print('    metadata.py copied')

# metadata_writers/ — fix all imports in one pass
MW_SRC = os.path.join(META_SRC, 'python', 'metadata_writers')
MW_DST = os.path.join(PKG, 'metadata_writers')
if os.path.exists(MW_DST):
    shutil.rmtree(MW_DST)
shutil.copytree(MW_SRC, MW_DST)
for root, _, files in os.walk(MW_DST):
    for fname in files:
        if not fname.endswith('.py'): continue
        fpath = os.path.join(root, fname)
        txt = open(fpath).read()
        txt = txt.replace('tensorflow_lite_support', 'tflite_support')
        open(fpath, 'w').write(txt)
print('    metadata_writers copied and fixed')

# ── Step 4: Activate shim ────────────────────────────────────────────────────
for k in list(sys.modules):
    if 'tflite' in k: del sys.modules[k]
if SHIM not in sys.path:
    sys.path.insert(0, SHIM)
importlib.invalidate_caches()

# ── Step 5: Smoke test ───────────────────────────────────────────────────────
print('[5] Testing...')
from tflite_support.metadata_writers import image_classifier, writer_utils
print('    SUCCESS: prerequisites ready')
```

## Create Model Metadata for Task Library and Codegen

<a name="image_classifiers"></a>
### Image classifiers

See the [image classifier model compatibility requirements](https://www.tensorflow.org/lite/inference_with_metadata/task_library/image_classifier#model_compatibility_requirements) for more details.

**Step 1:** Import the required packages.

```
from tflite_support.metadata_writers import image_classifier
from tflite_support.metadata_writers import writer_utils
```

**Step 2:** Download the example image classifier, [mobilenet_v2_1.0_224.tflite](https://github.com/tensorflow/tflite-support/blob/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/mobilenet_v2_1.0_224.tflite), and the [label file](https://github.com/tensorflow/tflite-support/blob/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/labels.txt).

```
!curl -L https://github.com/tensorflow/tflite-support/raw/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/mobilenet_v2_1.0_224.tflite -o mobilenet_v2_1.0_224.tflite
!curl -L https://github.com/tensorflow/tflite-support/raw/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/labels.txt -o mobilenet_labels.txt
```

**Step 3:** Create metadata writer and populate.

```
ImageClassifierWriter = image_classifier.MetadataWriter
_MODEL_PATH = "mobilenet_v2_1.0_224.tflite"
# Task Library expects label files that are in the same format as the one below.
_LABEL_FILE = "mobilenet_labels.txt"
_SAVE_TO_PATH = "mobilenet_v2_1.0_224_metadata.tflite"
# Normalization parameters are required when reprocessing the image. It is
# optional if the image pixel values are in the range of [0, 255] and the input
# tensor is quantized to uint8. See the introduction for normalization and
# quantization parameters below for more details.
# https://www.tensorflow.org/lite/models/convert/metadata#normalization_and_quantization_parameters)
_INPUT_NORM_MEAN = 127.5
_INPUT_NORM_STD = 127.5

# Create the metadata writer.
writer = ImageClassifierWriter.create_for_inference(
    writer_utils.load_file(_MODEL_PATH), [_INPUT_NORM_MEAN], [_INPUT_NORM_STD],
    [_LABEL_FILE])

# Verify the metadata generated by the metadata writer.
print(writer.get_metadata_json())

# Populate the metadata into the model.
writer_utils.save_file(writer.populate(), _SAVE_TO_PATH)
```

## Create Model Metadata with Semantic Information

You can fill in more descriptive information about the model and each tensor through the metadata writer API.

**Step 1:** Import the required packages.

```
from tflite_support.metadata_writers import image_classifier
from tflite_support.metadata_writers import metadata_info
from tflite_support.metadata_writers import writer_utils
from tflite_support import metadata_schema_py_generated as _metadata_fb
```

**Step 2:** Download the example image classifier (reuse from section above).

```
!curl -L https://github.com/tensorflow/tflite-support/raw/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/mobilenet_v2_1.0_224.tflite -o mobilenet_v2_1.0_224.tflite
!curl -L https://github.com/tensorflow/tflite-support/raw/master/tensorflow_lite_support/metadata/python/tests/testdata/image_classifier/labels.txt -o mobilenet_labels.txt
```

**Step 3:** Create model and tensor information.

```
model_buffer = writer_utils.load_file("mobilenet_v2_1.0_224.tflite")

# Create general model information.
general_md = metadata_info.GeneralMd(
    name="ImageClassifier",
    version="v1",
    description=("Identify the most prominent object in the image from a "
                 "known set of categories."),
    author="TensorFlow Lite",
    licenses="Apache License. Version 2.0")

# Create input tensor information.
input_md = metadata_info.InputImageTensorMd(
    name="input image",
    description=("Input image to be classified. The expected image is "
                 "128 x 128, with three channels (red, blue, and green) per "
                 "pixel. Each element in the tensor is a value between min and "
                 "max, where (per-channel) min is [0] and max is [255]."),
    norm_mean=[127.5],
    norm_std=[127.5],
    color_space_type=_metadata_fb.ColorSpaceType.RGB,
    tensor_type=writer_utils.get_input_tensor_types(model_buffer)[0])

# Create output tensor information.
output_md = metadata_info.ClassificationTensorMd(
    name="probability",
    description="Probabilities of the 1001 labels respectively.",
    label_files=[
        metadata_info.LabelFileMd(file_path="mobilenet_labels.txt",
                                  locale="en")
    ],
    tensor_type=writer_utils.get_output_tensor_types(model_buffer)[0])
```

**Step 4:** Create metadata writer and populate.

```
ImageClassifierWriter = image_classifier.MetadataWriter
_SAVE_TO_PATH = "mobilenet_v2_1.0_224_metadata.tflite"

# Create the metadata writer.
writer = ImageClassifierWriter.create_from_metadata_info(
    model_buffer, general_md, input_md, output_md)

# Verify the metadata generated by the metadata writer.
print(writer.get_metadata_json())

# Populate the metadata into the model.
writer_utils.save_file(writer.populate(), _SAVE_TO_PATH)
```

## Read the Metadata Populated to Your Model

You can display the metadata and associated files in a TFLite model through the `MetadataDisplayer`.

```
from tflite_support import metadata

displayer = metadata.MetadataDisplayer.with_model_file("mobilenet_v2_1.0_224_metadata.tflite")
print("Metadata populated:")
print(displayer.get_metadata_json())

print("Associated file(s) populated:")
for file_name in displayer.get_packed_associated_file_list():
  print("file name: ", file_name)
  print("file content:")
  print(displayer.get_associated_file_buffer(file_name))
```
